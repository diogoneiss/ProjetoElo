```python
import pandas as pd
```

```python
df = pd.read_csv("./brasileirao.csv")
```

```python
def drop_duplicates(original_df, csv_file_path="brasileirao.csv"):
    df = original_df.copy()
    df['HomeTeam_Duplicated'] = df['HomeTeam'].shift(1) == df['HomeTeam']
    df['AwayTeam_Duplicated'] = df['AwayTeam'].shift(1) == df['AwayTeam']

    duplicated_rows = df[(df['HomeTeam_Duplicated'] == True) & (df['AwayTeam_Duplicated'] == True)]
    
    df.drop(['HomeTeam_Duplicated', 'AwayTeam_Duplicated'], axis=1, inplace=True)

    # Save DataFrame to CSV
    df.to_csv(csv_file_path, index=False)

    return duplicated_rows

duplicated_rows = drop_duplicates(df)

duplicated_rows
```
